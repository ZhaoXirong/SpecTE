
## 老版本的星表需要分割，新版本不需要运行这个部分

```python
combined_name=df_lamost.columns[0]
combined_name
```

```python
# 使用 str.split() 方法按竖线分割数据列
new_columns = df_lamost[combined_name].str.split('|', expand=True)

# 为新列命名
# new_columns.columns = ['lamost_obsid', 'lamost_snrg', 'lamost_class', 'lamost_subclass', 'lamost_z', 'lamost_ra', 'lamost_dec', 'lamost_logg', 'lamost_file']
names = combined_name.split('|') # 使用split方法拆分字符串
names = [name.replace('combined', 'lamost') for name in names] # 替换前缀并创建新的列表
new_columns.columns = names

# 将新列添加到数据框中
df_lamost = pd.concat([df_lamost, new_columns], axis=1)

# 删除原始的单一列数据
df_lamost.drop(combined_name, axis=1, inplace=True)
```

## 按照snrg分组

```python
# 按照snrg分组

# (老版本需要)将 'lamost_snrg' 列转换为数值型，这里使用 pd.to_numeric，并忽略无法转换的错误
# df_lamost['lamost_snrg'] = pd.to_numeric(df_lamost['lamost_snrg'], errors='coerce')

# 分组：lamost_snrg < 50
group_1 = df_lamost[df_lamost['snrg'] < 50]

# 分组：lamost_snrg > 100
group_2 = df_lamost[df_lamost['snrg'] >= 100] 
group_2 = group_2[group_2['snrg']  < 200] 

# 分组：lamost_snrg > 100
group_3 = df_lamost[df_lamost['snrg'] >= 50] 
group_3 = group_3[group_3['snrg']  < 200] 
```

```python
group_1_path = r'G:\Star\1_Data_download_and_preprocessing\raw_data_catalogue_label\DR11_less_50.csv'
group_2_path = r'G:\Star\1_Data_download_and_preprocessing\raw_data_catalogue_label\DR11_between_100_200.csv'
group_3_path = r'G:\Star\1_Data_download_and_preprocessing\raw_data_catalogue_label\DR11_between_50_200.csv'


group_1.to_csv(group_1_path, index=False)
group_2.to_csv(group_2_path, index=False)
group_3.to_csv(group_3_path, index=False)
```
